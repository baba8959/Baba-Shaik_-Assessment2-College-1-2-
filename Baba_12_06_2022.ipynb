```python
def varaince(data):
  n = len(data)
  mean = sum(data) / n
  devition = [(x - mean) ** 2 for x in data]
  varaince = sum(devition) /  n
  return varaince
```

```python
varaince ([1,2,3,4,5])
```

```python
import math 

def stdev(data):
    var = varaince(data)
    std_dev = math.sqrt(var)
    return(std_dev)
```

```python
stdev([1,2,3,4,5])
```

```python
# obtain simple random sample
simple_random_sample = df.sample(n=5).sort_values(by='product_id')
#save the sample mean in a separate variable 
simple_random_mean = round(simple_random_sample['measure'].mean(),3)
#view sampled data frame
simple_random_sample
```

systematic sampling

```python
# define systematic sampling function
def systematic_sampling(df,step):
     indexs = np.arange(0,len(df),step = step)
     systematic_sample = df.iloc[indexes]
     return systematic_sample
# obtain a systematic sample and save it in a new variable
systematic_sample = systematic_sampling (df, 4)
# save the sample mean in a separate variable
systematic_mean = round(systematic_sample['measure'].mean(),3)
# view sampled data frame
systematic_sample
```

```python
plt.title('box plot')
sb.boxplot (y = 'sepal_width',x = 'species', data = df)

sb.set (style = 'whitegrid', font_scale = 1)
plt.show()
```
